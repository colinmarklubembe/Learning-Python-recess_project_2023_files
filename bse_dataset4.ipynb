```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#control display of seaborn data
pd.options.display.max_columns = 50
sns.set(style='darkgrid', rc={"axes.facecolor": "#CAF1DE"})
```

```python
#load the dataset
df_4 = pd.read_csv('datasets/bse_dataset4.csv')
df_4
```

```python
#shape of the dataset
df_4.shape
```

```python
#colums of the dataset
df_4.columns
```

```python
#drop empty columns
df_4.dropna(inplace=True)
df_4.shape
```

```python
#using fillna() to fill empty values
df_4['Calories'].fillna(value = 100, inplace=True)
df_4.shape
```

```python
print(df_4.to_string())
```

```python
#to fill in the mean value of the column
df_4['Calories'].fillna(value = df_4['Calories'].mean(), inplace=True)
df_4.shape
```

```python
#for the Age column
m = df_4['Age'].mean()
df_4['Age'].fillna(value = m, inplace=True)
print(df_4.to_string())
```

```python
# Return rows with the mean value of the column Age
df_4[df_4['Age'] == m]
```

```python
# Set the option to display all rows
pd.set_option('display.max_rows', None)
```

```python
#calculate median
med = df_4['Calories'].median()
print(med)
df_4['Calories'].fillna(value = med, inplace=True)
print(df_4)
```

```python
#calculating the mode on age
mode = df_4['Age'].mode()
print(mode)
df_4['Age'].fillna(value = mode, inplace=True)
print(df_4)
```

```python
#wrong formats in rows that need to be removed
df_4['Timestamp'].fillna("08/07/2023 09:45:13", inplace=True)
print(df_4['Timestamp'])
```

```python
#check if format is correct
df_4['Timestamp'] = pd.to_datetime(df_4['Timestamp'])
print(df_4['Timestamp'])
```

```python
# dealing with wrong data
for x in df_4.index:
    if df_4.loc[x, 'gender'] == "Marabou stock":
        df_4.drop(x, inplace=True)
print(df_4['gender'])
df_4.shape
```

```python
#duplicates
print(df_4.duplicated())
df_4.drop_duplicates(inplace= True)
```

```python
print(df_4.duplicated())
```
